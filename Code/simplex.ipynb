```python
# -*- coding: utf-8 -*-

"""

Created on Sat Dec 11 07:48:23 2021



@author: chane

"""



import numpy as np

from fractions import Fraction # Affichage des nombres entier



print("\n				 ****SiMplex Algorithm ****\n\n")



# Les entrées



# Valeurs des coefficients des contraintes

A = np.array([[1, 1, 0, 1], [2, 1, 1, 0]])

# b contient les valeurs 

b = np.array([8, 10])		

# c contient les valeurs objective de la fonction objective Z	

c = np.array([1, 1, 0, 0])			



# B contient les valeurs de la variable basic identity matrix

cb = np.array(c[3])

B = np.array([[3], [2]])		

# cb contient des coefficients in Z

cb = np.vstack((cb, c[2]))		

xb = np.transpose([b])				

# combinaison de la matrice B and cb

table = np.hstack((B, cb))			

table = np.hstack((table, xb))		

# combinaison de matrice B, cb and xb

# Combinaison complète de la matrix A to de la forme complete de simplex 

table = np.hstack((table, A))		

# Changer la valeur de la table en type float

table = np.array(table, dtype ='float')

# entrer et fin



# si min problem, make this var 1

MIN = 0



print("Table at itr = 0")

print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")

for row in table:

	for el in row:

				# limiter la valeur de dénominateur à  100

		print(Fraction(str(el)).limit_denominator(100), end ='\t')

	print()

print()

print("Simplex....")



# when optimality reached it will be made 1

reached = 0	

itr = 1

unbounded = 0

alternate = 0



while reached == 0:



	print("Iteration: ", end =' ')

	print(itr)

	print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")

	for row in table:

		for el in row:

			print(Fraction(str(el)).limit_denominator(100), end ='\t')

		print()



	# calcul cj - zj pour les valeurs non-basics

	i = 0

	rel_prof = []

	while i<len(A[0]):

		rel_prof.append(c[i] - np.sum(table[:, 1]*table[:, 3 + i]))

		i = i + 1



	print("rel profit: ", end =" ")

	for profit in rel_prof:

		print(Fraction(str(profit)).limit_denominator(100), end =", ")

	print()

	i = 0

	

	b_var = table[:, 0]

	# chercher une solution alternative

	while i<len(A[0]):

		j = 0

		present = 0

		while j<len(b_var):

			if int(b_var[j]) == i:

				present = 1

				break;

			j+= 1

		if present == 0:

			if rel_prof[i] == 0:

				alternate = 1

				print("Case of Alternate found")

				# afficher(i, end =" ")

		i+= 1

	print()

	flag = 0

	for profit in rel_prof:

		if profit>0:

			flag = 1

			break

		# si tout les valeurs sont relatives  <= 0

	if flag == 0:

		print("All profits are <= 0, optimality reached")

		reached = 1

		break



	# kth var will enter les valeurs basics

	k = rel_prof.index(max(rel_prof))

	min = 99999

	i = 0;

	r = -1

	# min ratio test (sur les valeurs positives )

	while i<len(table):

		if (table[:, 2][i]>0 and table[:, 3 + k][i]>0):

			val = table[:, 2][i]/table[:, 3 + k][i]

			if val<min:

				min = val

				r = i	 # leaving variable

		i+= 1



		# sinon on fait le test de performance

	if r ==-1:

		unbounded = 1

		print("Case of Unbounded")

		break



	print("chercher la valeur de l'element de pivot :", end =' ')

	print(np.array([r, 3 + k]))



	pivot = table[r][3 + k]

	print("valeur de pivot : ", end =" ")

	print(Fraction(pivot).limit_denominator(100))

		

		# les opérations sur les lignes

	# diviser la ligne de pivot avec l'élément de pivot

	table[r, 2:len(table[0])] = table[

			r, 2:len(table[0])] / pivot

			

	# faire sur les lignes les operations sur n'importe quelle autre ligne

	i = 0

	while i<len(table):

		if i != r:

			table[i, 2:len(table[0])] = table[i,

				2:len(table[0])] - table[i][3 + k] 

			table[r, 2:len(table[0])]

		i += 1



	

	# affecter la nouvelle valeur de la variable basic

	table[r][0] = k

	table[r][1] = c[k]

	

	print()

	print()

	itr+= 1

	



print()



print("***************************************************************")

if unbounded == 1:

	print("UNBOUNDED LPP")

	exit()

if alternate == 1:

	print("Solution Permanante")



print("table optimal :")

print("B \tCB \tXB \ty1 \ty2 \ty3 \ty4")

for row in table:

	for el in row:

		print(Fraction(str(el)).limit_denominator(100), end ='\t')

	print()

print()

print("valeur de Z optimiser: ", end =" ")



basis = []

i = 0

sum = 0

while i<len(table):

	sum += c[int(table[i][0])]*table[i][2]

	temp = "x"+str(int(table[i][0])+1)

	basis.append(temp)

	i+= 1

# si c'est un problème de MIN on multiplie z  par negative

if MIN == 1:

	print(-Fraction(str(sum)).limit_denominator(100))

else:

	print(Fraction(str(sum)).limit_denominator(100))

print("Valeur de variable Basic finale: ", end =" ")

print(basis)



print("Fin de Simplex ")

print()
```
